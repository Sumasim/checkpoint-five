## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb
import datetime as dt

shl_df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
shl_df.head()
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
shl_df.info()
```

```python
shl_df.describe()
```

```python
correlation_matrix = shl_df.corr()
plt.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, cmap="coolwarm", annot=True)
plt.title("Correlation Heatmap of Health Metrics")
plt.show()
```

```python
selected_columns = ["Age", "Sleep Duration", "Physical Activity Level", "Stress Level", "Heart Rate"]
sb.pairplot(shl_df[selected_columns])
plt.suptitle("Pair Plot of Health Metrics", y=1.02)
plt.show()
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
X = shl_df['Age']
y = shl_df['Stress Level']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
X_train
y_train
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_sm).fit()

lr.params
lr.summary()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 10.4310 + 13.8817*X_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
y_train_pred = lr.predict(X_train_sm)

# Creating residuals from the Y_train data and predicted Y_data
res = (y_train - y_train_pred)
```

```python
fig = plt.figure()
sb.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
